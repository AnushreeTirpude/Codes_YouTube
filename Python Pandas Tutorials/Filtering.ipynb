# Filtering 

```python
import pandas as pd     #importing pandas
```

```python
df = pd.DataFrame({'A':[1,5,8,5],'B':[5,6,8,10],'C':[4,8,'python',9+5j]})
```

```python
df
```

```python
filt = (df.loc[2] == 8)
filt 
```

```python
filt=(df['C']=='python')
```

```python
df[filt]
```

```python
df.loc[filt]    
```

```python
df.loc[df['C']=='python','C']         #can get specific column as well
```

```python
filter=(df['C']=='python') | (df['A']==1)       #can use for filtering multiple conditions as well using and , or
```

```python
filter
```

```python
df.loc[filter]
```

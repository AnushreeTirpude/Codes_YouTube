Ques) What is the maximum number of Total Confirmed cases in each state?

```python
grp=df.groupby('Name of State / UT')
print(grp['Total Confirmed cases'].agg(max))
```

```python
#Grouping on multiple columns
```

```python
df.groupby(['Date'])['Death','Total Confirmed cases','Cured/Discharged/Migrated','Name of State / UT'].max()          #groupby Date and have columns against that date, get the maximum cases on that date
```

```python
covid_perday=df.groupby(['Date'])['Death','Total Confirmed cases','Cured/Discharged/Migrated'].max()
```

```python
covid_perday['Total Confirmed cases'].max()               #max no of cases
```

```python
covid_perday['Total Confirmed cases'].idxmax()    
```

Ques) What is sum of total number of confirmed cases in each state?

```python
grp=df.groupby('Name of State / UT')
print(grp['Total Confirmed cases'].agg(sum))
```

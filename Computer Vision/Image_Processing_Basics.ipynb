## Basic Image Operations

### Save and Resize an image

Resizing images

```python
# Resize
resized = cv2.resize(image, (300,300))

print(image.shape)
print(resized.shape)
```

```python
# Change color space to Gray
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Showing the image
cv2.imshow('image',image)
cv2.waitKey()  
cv2.destroyAllWindows()
```

### Crop an Image

```python
height, width = image.shape[0:2]
startRow = int(height*.15)
startCol = int(width*.15)
endRow = int(height*.85)
endCol = int(width*.85)
croppedImage = image[startRow:endRow, startCol:endCol]
cv2.imshow('Original Image', image)
cv2.imshow('Cropped Image', croppedImage)
cv2.waitKey(0)
```

## Image rotation

To get the rotation matrix, we use the getRotationMatrix2D() method.
Its syntax is:
cv2.getRotationMatrix2D(center, angle, scale)
where center is the center point of rotation, the angle is the angle in degrees and scale is the scale property which makes the image fit on the screen

```python
height, width = image.shape[0:2]
rotationMatrix = cv2.getRotationMatrix2D((width/2, height/2), 90, .5)
rotatedImage = cv2.warpAffine(image, rotationMatrix, (width, height))
cv2.imshow('Rotated Image', rotatedImage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
